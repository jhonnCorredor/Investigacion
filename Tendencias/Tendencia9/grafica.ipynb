```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Libro1.csv', delimiter=';')

# Convertir la columna 'Semana' a formato de fecha
data['Semana'] = pd.to_datetime(data['Semana'], format='%d/%m/%Y')

# Configurar el gráfico
plt.figure(figsize=(12, 8))

# Graficar cada columna de interés
plt.plot(data['Semana'], data['java: (Todo el mundo)'], label='Java', marker='o')
plt.plot(data['Semana'], data['C++: (Todo el mundo)'], label='C++', marker='o')
plt.plot(data['Semana'], data['python: (Todo el mundo)'], label='Python', marker='o')
plt.plot(data['Semana'], data['c#: (Todo el mundo)'], label='C#', marker='o')
plt.plot(data['Semana'], data['javaScript: (Todo el mundo)'], label='JavaScript', marker='o')

# Configurar los detalles del gráfico
plt.xlabel('Semana')
plt.ylabel('Puntuación')
plt.title('Puntuaciones Semanales por Lenguaje de Programación')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
