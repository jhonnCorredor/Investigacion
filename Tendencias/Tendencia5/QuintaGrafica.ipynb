```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV
archivo_csv = "QuintaTendencia.csv"  # Cambia por la ruta real de tu archivo
df = pd.read_csv(archivo_csv, sep=';')

# Limpiar los datos
df = df.dropna(how='all')  # Eliminar filas completamente vac√≠as

# Mostrar las primeras filas para verificar
df.head()
```

```python
# Crear un gr√°fico de barras para el salario base mediano
plt.figure(figsize=(12, 8))  # Tama√±o del gr√°fico

# Crear el gr√°fico de barras
plt.barh(df['Job Title'], df['Median Base Salary'], color='skyblue')

# A√±adir etiquetas y t√≠tulo
plt.xlabel('Salario Base Mediano')
plt.ylabel('T√≠tulo del Trabajo')
plt.title('Salario Base Mediano por T√≠tulo de Trabajo')

# Mostrar la gr√°fica
plt.show()
```

```python
# Crear un gr√°fico de barras para el n√∫mero de aperturas de empleos
plt.figure(figsize=(12, 8))  # Tama√±o del gr√°fico

# Crear el gr√°fico de barras
plt.barh(df['Job Title'], df['Job Openings'], color='salmon')

# A√±adir etiquetas y t√≠tulo
plt.xlabel('N√∫mero de Aperturas de Empleos')
plt.ylabel('T√≠tulo del Trabajo')
plt.title('N√∫mero de Aperturas de Empleos por T√≠tulo de Trabajo')

# Mostrar la gr√°fica
plt.show()
```
