```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV
archivo_csv = "primerTendencia.csv"  # Cambia por la ruta real de tu archivo
df = pd.read_csv(archivo_csv, sep=';')

# Mostrar las primeras filas del archivo para verificar los datos
df.head()
```

```python
df.plot(kind='bar', x='Category', y=['2023 Spending', '2024 Spending'], color=['blue', 'green'])

# Añadir etiquetas y título
plt.xlabel('Categoría')
plt.ylabel('Gasto (USD)')
plt.title('Comparación de gastos IT en 2023 vs 2024')

# Mostrar la gráfica
plt.show()
```

```python
# Crear una gráfica de barras para comparar el crecimiento porcentual de 2023 y 2024
df.plot(kind='bar', x='Category', y=['2023 Growth (%)', '2024 Growth (%)'], color=['red', 'orange'])

# Añadir etiquetas y título
plt.xlabel('Categoría')
plt.ylabel('Crecimiento (%)')
plt.title('Comparación de crecimiento IT en 2023 vs 2024')

# Mostrar la gráfica
plt.show()
```
