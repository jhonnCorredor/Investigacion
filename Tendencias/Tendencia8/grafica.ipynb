```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Intentar leer el archivo CSV con varias codificaciones
df = pd.read_excel('DesafiosEmpresas.csv', sheet_name='Hoja1')

# Crear el grfico
plt.figure(figsize=(10, 6))
plt.bar(df['Challenge'], df['Percentage'], color='skyblue')

# Configurar el grfico
plt.xlabel('Desafio')
plt.ylabel('Porcentaje')
plt.title('Desafios Empresas')
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el grfico
plt.show()
```
