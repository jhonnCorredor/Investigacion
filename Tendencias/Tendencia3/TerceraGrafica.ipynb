```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV
archivo_csv = "TerceraTendencia.csv"  # Cambia por la ruta real de tu archivo
df = pd.read_csv(archivo_csv, sep=';')

# Separar la fila de "Prob. Adopción"
df_prob_adopcion = df[df['Año'] == 'Prob. Adopción']

# Filtrar el resto de las filas para excluir "Prob. Adopción"
df_filtrado = df[df['Año'] != 'Prob. Adopción']

# Convertir los porcentajes en números flotantes y remover el símbolo '%'
df_filtrado = df_filtrado.replace('%', '', regex=True).replace('-', '0', regex=True).astype(float)

# Mostrar las primeras filas para verificar
df_filtrado.head()

```

```python
# Definir los lenguajes que queremos graficar
lenguajes = ['JavaScript', 'Python', 'HTML/CSS', 'SQL', 'Java']

# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))  # Ajustar el tamaño del gráfico
for lenguaje in lenguajes:
    plt.plot(df_filtrado['Año'], df_filtrado[lenguaje], label=lenguaje)

# Añadir etiquetas y título
plt.xlabel('Año')
plt.ylabel('Porcentaje de Uso')
plt.title('Evolución del uso de lenguajes de programación (2017-2023)')
plt.legend()

# Mostrar la gráfica
plt.show()
```
