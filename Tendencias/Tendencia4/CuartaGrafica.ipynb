```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV
archivo_csv = "CuartaTendencia.csv"  # Cambia por la ruta real de tu archivo
df = pd.read_csv(archivo_csv, sep=';')

# Limpiar los datos
df = df.dropna(how='all')  # Eliminar filas completamente vacías
df = df.iloc[:, :2]  # Seleccionar solo las primeras dos columnas relevantes

# Renombrar las columnas para mayor claridad
df.columns = ['Technology', 'Percentage']

# Convertir la columna 'Percentage' a tipo numérico, eliminando el símbolo '%' si está presente
df['Percentage'] = df['Percentage'].replace('', '0').astype(float)

# Mostrar las primeras filas del archivo para verificar
df.head()
```

```python
# Crear la gráfica de barras
plt.figure(figsize=(12, 8))  # Tamaño del gráfico

# Ordenar los datos por porcentaje para mejor visualización
df_sorted = df.sort_values(by='Percentage', ascending=False)

# Crear el gráfico de barras
plt.barh(df_sorted['Technology'], df_sorted['Percentage'], color='skyblue')

# Añadir etiquetas y título
plt.xlabel('Porcentaje')
plt.ylabel('Tecnología')
plt.title('Porcentaje de Uso de Tecnologías')

# Mostrar la gráfica
plt.show()
```
