```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV
archivo_csv = "SegundaTendencia.csv"  # Cambia por la ruta real de tu archivo
df = pd.read_csv(archivo_csv, sep=';')

# Convertir los porcentajes a n√∫meros flotantes
df['Porcentaje'] = df['Porcentaje'].str.replace('%', '').astype(float)

# Mostrar las primeras filas del archivo para verificar los datos
df.head()
```

```python
# Crear una gr√°fica de pastel
plt.figure(figsize=(8, 8))  # Ajustar el tama√±o de la figura

plt.pie(df['Porcentaje'], labels=df['Lenguaje'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# A√±adir un t√≠tulo
plt.title('Porcentaje de uso de lenguajes de programaci√≥n')

# Mostrar la gr√°fica
plt.show()
```
