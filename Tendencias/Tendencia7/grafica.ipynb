```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV
archivo_csv = "SeptimaTendencia.csv"  # Cambia por la ruta real de tu archivo
df = pd.read_csv(archivo_csv, sep=';')

# Mostrar las primeras filas del DataFrame para verificar
print(df.head())

# Verificar el tipo de datos de la columna 'Ratings'
print(df['Ratings'].dtype)

# Convertir la columna 'Ratings' a tipo string y luego limpiar y convertir a float
df['Ratings'] = df['Ratings'].astype(str).str.replace('%', '', regex=False).astype(float)

# Mostrar el DataFrame para verificar cambios
print(df.head())
```

```python
# Crear el grfico de barras para las calificaciones de los lenguajes de programacin
plt.figure(figsize=(12, 8))  # Tamao del grfico

# Crear la grfica de barras usando los datos de calificaciones
plt.barh(df['Programming Language'], df['Ratings'], color='teal')

# Aadir etiquetas y ttulo
plt.xlabel('Calificacin (%)')
plt.ylabel('Lenguaje de Programacin')
plt.title('Calificacin de Lenguajes de Programacin en Sep-24')

# Mostrar la grfica
plt.show()
```
