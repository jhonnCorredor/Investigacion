```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Intentar leer el archivo CSV con diferentes codificaciones
df = pd.read_excel('Software_Development_Tools.xlsx', sheet_name='Hoja1')

# Crear el grfico
plt.figure(figsize=(10, 6))
plt.bar(df['Categora'], df['Porcentaje'], color='skyblue')

# Configurar el grfico
plt.xlabel('Categora')
plt.ylabel('Porcentaje')
plt.title('Porcentaje por Categora')
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el grfico
plt.show()
```
