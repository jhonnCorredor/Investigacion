```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Intentar leer el archivo CSV con diferentes codificaciones
df = pd.read_excel('Software_Development_Tools.xlsx', sheet_name='Hoja1')

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.bar(df['Categoría'], df['Porcentaje'], color='skyblue')

# Configurar el gráfico
plt.xlabel('Categoría')
plt.ylabel('Porcentaje')
plt.title('Porcentaje por Categoría')
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
